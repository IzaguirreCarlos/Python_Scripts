```python
import tkinter as tk
from tkinter import messagebox
from urllib.request import urlopen
from urllib.error import URLError, HTTPError

def verificar_conectividad():
    """
    Función para verificar la conectividad de un sitio web ingresado por el usuario.
    """
    url = entrada_url.get()  # Obtener la URL ingresada por el usuario

    # Asegurarse de que la URL comience con http:// o https://
    if not url.startswith("http://") and not url.startswith("https://"):
        url = "http://" + url

    try:
        # Intentar abrir la URL y obtener el código de estado HTTP
        respuesta = urlopen(url)
        codigo_estado = respuesta.getcode()

        # Si el código de estado es 200, el sitio es accesible
        if codigo_estado == 200:
            messagebox.showinfo("Éxito", f"El sitio web '{url}' es accesible.")
        else:
            messagebox.showwarning("Advertencia", f"El sitio web '{url}' devolvió el código de estado {codigo_estado}.")

    except HTTPError as e:
        # Manejar errores HTTP (por ejemplo, 404, 500, etc.)
        messagebox.showerror("Error", f"Error HTTP: {e.code} - {e.reason}")

    except URLError as e:
        # Manejar errores de URL (por ejemplo, dominio inválido, problemas de conexión)
        messagebox.showerror("Error", f"Error de URL: {e.reason}")

    except Exception as e:
        # Capturar cualquier otra excepción
        messagebox.showerror("Error", f"Ocurrió un error inesperado: {str(e)}")

def crear_gui():
    """
    Función para crear la interfaz gráfica de usuario (GUI) para el verificador de conectividad.
    """
    # Crear la ventana principal de la aplicación
    root = tk.Tk()
    root.title("Verificador de Conectividad de Sitios Web")
    root.geometry("400x200")

    # Etiqueta para instruir al usuario
    etiqueta_instrucciones = tk.Label(root, text="Ingresa la URL del sitio web:", font=("Arial", 12))
    etiqueta_instrucciones.pack(pady=10)

    # Widget de entrada para la URL
    global entrada_url
    entrada_url = tk.Entry(root, width=50, font=("Arial", 12))
    entrada_url.pack(pady=10)

    # Botón para verificar la conectividad
    boton_verificar = tk.Button(root, text="Verificar Conectividad", command=verificar_conectividad, font=("Arial", 12))
    boton_verificar.pack(pady=10)

    # Ejecutar el bucle de eventos de Tkinter
    root.mainloop()

# Ejecutar la GUI
if __name__ == "__main__":
    crear_gui()


# https://www.google.com 

# "https:sitioinexistente12345.com
```
