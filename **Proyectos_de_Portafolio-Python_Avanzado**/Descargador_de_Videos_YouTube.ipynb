DESCARGADOR DE VIDEOS DE YOUTUBE :

En este proyecto, crearás una aplicación sencilla con una interfaz de usuario para descargar vídeos de YouTube en el formato y nivel de calidad deseados. En esta aplicación, debes proporcionar el enlace al vídeo de YouTube que necesitas descargar en tu máquina.

Al practicar este proyecto, aprenderá a crear una ventana GUI utilizando casillas de verificación, cuadros de entrada y botones. En este proyecto, utilizará los módulos Pytube y Tkinter para crear esta aplicación. El módulo Pytube ayuda a las personas a descargar videos de YouTube.

```python

import tkinter as tk
from tkinter import Label, Entry, Checkbutton, Button, Tk, StringVar, IntVar
from pytube import YouTube
from pytube.exceptions import VideoUnavailable

def obtener_resoluciones(video):
    """Obtiene las resoluciones disponibles para el video."""
    streams = video.streams.filter(file_extension='mp4', progressive=True)
    resoluciones = {stream.itag: stream.resolution for stream in streams}
    return resoluciones

def actualizar_opciones():
    """Actualiza las casillas de verificación con resoluciones disponibles."""
    enlace_video = enlace.get()
    try:
        video = YouTube(enlace_video)
        resoluciones = obtener_resoluciones(video)

        # Limpiar opciones anteriores
        for widget in marco_resoluciones.winfo_children():
            widget.destroy()

        # Crear nuevas opciones
        for itag, resolucion in resoluciones.items():
            Checkbutton(marco_resoluciones, text=resolucion, variable=res_vars[itag]).pack(anchor='w')

    except Exception as e:
        Label(marco_resoluciones, text=f"Error: {e}", fg="red").pack()

def descargar_video():
    enlace_video = enlace.get()
    try:
        video = YouTube(enlace_video)
        for itag, var in res_vars.items():
            if var.get():
                video_stream = video.streams.get_by_itag(itag)
                video_stream.download(output_path="video_path", filename="video_descargado.mp4")
                Label(ventana, text="Descarga completada con éxito", fg="green").pack()
                return

        Label(ventana, text="Por favor, selecciona una resolución", fg="red").pack()
    except Exception as e:
        Label(ventana, text=f"Error: {e}", fg="red").pack()

# Crear la ventana principal
ventana = Tk()
ventana.geometry("700x500")
ventana.title("Descargador de Videos de YouTube")

# Crear etiquetas y campos de entrada
Label(ventana, text="DESCARGADOR DE VIDEOS DE YOUTUBE", bg='grey', font=('Calibri', 15)).pack(pady=10)
Label(ventana, text="Introduce el enlace para descargar", font=('Calibri', 12)).pack(pady=5)

enlace = StringVar()
Entry(ventana, textvariable=enlace, width=50).pack(pady=5)

Button(ventana, text="Actualizar Opciones", bg='blue', command=actualizar_opciones).pack(pady=10)

# Crear marco para resoluciones disponibles
marco_resoluciones = tk.Frame(ventana)
marco_resoluciones.pack(pady=10)

res_vars = {}

# Crear botón para iniciar la descarga
Button(ventana, text="Descargar", bg='green', command=descargar_video).pack(pady=20)

# Ejecutar la ventana principal
ventana.mainloop()

# https://www.youtube.com/watch?v=YhvLXwmCO4E


# https://www.youtube.com/watch?v=LI7NEs_Y5gg
 
#  https://www.youtube.com/watch?v=eFPr6J-ggdg
```
