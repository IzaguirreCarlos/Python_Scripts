APLICACION DE PRUEBA CUESTIONARIO :

En este proyecto, desarrollará una aplicación de cuestionario que solicitará a los usuarios que respondan una serie de preguntas y luego mostrará las respuestas correctas. La aplicación muestra la puntuación de cada prueba. Esta aplicación permite la creación de nuevas cuentas de usuario. Puede asignar a los usuarios seleccionados como administradores en la aplicación. Los administradores pueden agregar preguntas para otros usuarios.

```python
# JSON es una sintaxis para almacenar e intercambiar datos.
# Python tiene un paquete integrado llamado json, que se utiliza para trabajar con datos JSON.
import json
# El módulo Random de Python es un módulo integrado que se utiliza para generar números aleatorios.
import random
# Módulo para entradas sin eco (las contraseñas no se muestran).
import getpass

# Variable global para almacenar datos del usuario autenticado
usuario = []

# Función para jugar el quiz
def jugar():
    print("\n==========INICIO DEL QUIZ==========")
    puntuacion = 0
    with open("assets/questions.json", 'r+') as archivo:
        preguntas = json.load(archivo)
        for i in range(10):
            total_preguntas = len(preguntas)
            indice = random.randint(0, total_preguntas - 1)
            print(f'\nPREGUNTA {i+1}: {preguntas[indice]["question"]}\n')
            for opcion in preguntas[indice]["options"]:
                print(opcion)
            respuesta = input("\nIntroduce tu respuesta: ")
            if preguntas[indice]["answer"][0] == respuesta[0].upper():
                print("\nRespuesta correcta")
                puntuacion += 1
            else:
                print("\nRespuesta incorrecta")
            del preguntas[indice]  # Eliminar la pregunta respondida
        print(f'\nPUNTUACIÓN FINAL: {puntuacion}')

# Función para agregar preguntas al quiz
def agregar_preguntas():
    if len(usuario) == 0:
        print("Debes iniciar sesión antes de agregar preguntas.")
    elif len(usuario) == 2:
        if usuario[1] == "ADMIN":
            print('\n==========AGREGAR PREGUNTAS==========\n')
            pregunta = input("Introduce la pregunta que deseas agregar:\n")
            opciones = []
            print("Introduce las 4 opciones con iniciales (A, B, C, D)")
            for _ in range(4):
                opciones.append(input())
            respuesta = input("Introduce la respuesta correcta:\n")
            with open("assets/questions.json", 'r+') as archivo:
                preguntas = json.load(archivo)
                nueva_pregunta = {"question": pregunta, "options": opciones, "answer": respuesta}
                preguntas.append(nueva_pregunta)
                archivo.seek(0)
                json.dump(preguntas, archivo)
                archivo.truncate()
                print("Pregunta agregada exitosamente.")
        else:
            print("No tienes permiso para agregar preguntas. Solo los administradores pueden hacerlo.")

# Función para crear una cuenta nueva
def crear_cuenta():
    print("\n==========CREAR CUENTA==========")
    nombre_usuario = input("Introduce tu NOMBRE DE USUARIO: ")
    contrasena = getpass.getpass(prompt='Introduce tu CONTRASEÑA: ')
    with open('assets/user_accounts.json', 'r+') as archivo_usuarios:
        usuarios = json.load(archivo_usuarios)
        if nombre_usuario in usuarios.keys():
            print("Ya existe una cuenta con este nombre de usuario.\nPor favor, inicia sesión.")
        else:
            usuarios[nombre_usuario] = [contrasena, "PLAYER"]
            archivo_usuarios.seek(0)
            json.dump(usuarios, archivo_usuarios)
            archivo_usuarios.truncate()
            print("¡Cuenta creada exitosamente!")

# Función para iniciar sesión
def iniciar_sesion():
    print('\n==========PANEL DE INICIO DE SESIÓN==========')
    nombre_usuario = input("NOMBRE DE USUARIO: ")
    contrasena = getpass.getpass(prompt='CONTRASEÑA: ')
    with open('assets/user_accounts.json', 'r') as archivo_usuarios:
        usuarios = json.load(archivo_usuarios)
    if nombre_usuario not in usuarios.keys():
        print("No existe una cuenta con ese nombre.\nPor favor, crea una cuenta primero.")
    elif usuarios[nombre_usuario][0] != contrasena:
        print("La contraseña es incorrecta.\nPor favor, inténtalo de nuevo.")
    else:
        print("Has iniciado sesión exitosamente.\n")
        usuario.append(nombre_usuario)
        usuario.append(usuarios[nombre_usuario][1])

# Función para cerrar sesión
def cerrar_sesion():
    global usuario
    if len(usuario) == 0:
        print("Ya estás desconectado.")
    else:
        usuario = []
        print("Te has desconectado exitosamente.")

# Función para mostrar las reglas del juego
def mostrar_reglas():
    print('''\n==========REGLAS==========
1. Cada ronda consta de 10 preguntas aleatorias. Para responder, debes presionar A/B/C/D (no importa mayúsculas o minúsculas).
2. Cada pregunta vale 1 punto. No hay puntos negativos por respuestas incorrectas.
3. Puedes crear una cuenta desde el panel de CREACIÓN DE CUENTAS.
4. Debes iniciar sesión desde el PANEL DE INICIO DE SESIÓN para jugar o agregar preguntas.
    ''')

# Función para mostrar información sobre el proyecto
def acerca_de():
    print('''\n==========ACERCA DE==========
Este proyecto ha sido creado por CARLOS IZAGUIRRE
Es un proyecto  de Python Avanzado para mi Portafolio de Proyectos es una
APLICACION  de Examen de Preguntas y Respouesta con Los . Resultados  acertados.''')

# Punto de entrada principal del programa
if __name__ == "__main__":
    opcion = 1
    while opcion != 7:
        print('\n=========BIENVENIDO A Preguntas De La Biblia MASTERS_King PALABRA MIEL==========')
        print('-----------------------------------------')
        print('1. JUGAR QUIZ')
        print('2. AGREGAR PREGUNTAS AL QUIZ')
        print('3. CREAR UNA CUENTA')
        print('4. INICIAR SESIÓN')
        print('5. CERRAR SESIÓN')
        print('6. VER REGLAS DEL JUEGO')
        print('7. SALIR')
        print('8. ACERCA DE')
        opcion = int(input('INTRODUCE TU OPCIÓN: '))
        if opcion == 1:
            jugar()
        elif opcion == 2:
            agregar_preguntas()
        elif opcion == 3:
            crear_cuenta()
        elif opcion == 4:
            iniciar_sesion()
        elif opcion == 5:
            cerrar_sesion()
        elif opcion == 6:
            mostrar_reglas()
        elif opcion == 7:
            break
        elif opcion == 8:
            acerca_de()
        else:
            print('OPCIÓN INCORRECTA. INTRODUCE TU ELECCIÓN NUEVAMENTE.')
```
