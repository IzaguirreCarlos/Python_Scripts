En este proyecto, creará la interfaz gráfica de usuario de la calculadora. Diseñará botones para realizar diversas operaciones y mostrar los resultados en la pantalla. Agregará funciones aritméticas esenciales además de los cálculos matemáticos avanzados.

Al practicar este proyecto, aprenderá a utilizar funciones predefinidas, funciones definidas por el usuario y formato de cadenas en el desarrollo de aplicaciones Python. Utilizará el módulo Tkinter para crear esta aplicación

```python
import tkinter as tk
from tkinter import messagebox

# Clase para gestionar las operaciones matemáticas
class Calculadora:
    def __init__(self):
        self.expresion = ""

    def agregar(self, valor):
        """Agrega un valor a la expresión actual."""
        self.expresion += str(valor)

    def limpiar(self):
        """Limpia la expresión actual."""
        self.expresion = ""

    def calcular(self):
        """Evalúa la expresión actual y devuelve el resultado."""
        try:
            resultado = eval(self.expresion)
            return resultado
        except Exception as e:
            return "Error"

# Clase para gestionar la interfaz gráfica
class InterfazCalculadora:
    def __init__(self, root):
        self.root = root
        self.root.title("Calculadora con Intrefaz GUL Carlos Desarrollador")
        self.root.geometry("500x600")  # Establece un tamaño más grande para la ventana
        self.calculadora = Calculadora()
        self.crear_interfaz()

    def crear_interfaz(self):
        """Crea los elementos de la interfaz gráfica."""
        # Pantalla de la calculadora
        self.pantalla = tk.Entry(self.root, font=("Arial", 24), borderwidth=5, relief=tk.RIDGE, justify=tk.RIGHT)
        self.pantalla.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

        # Botones de la calculadora
        botones = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('+', 4, 2), ('=', 4, 3),
            ('C', 5, 0)
        ]

        for (texto, fila, columna) in botones:
            if texto in ['+', '-', '*', '/']:
                boton = tk.Button(
                    self.root, 
                    text=texto, 
                    width=10, 
                    height=2, 
                    bg='orange',  # Cambia el color de fondo a anaranjado
                    command=lambda t=texto: self.agregar_a_pantalla(t)
                )
            elif texto == '=':
                boton = tk.Button(
                    self.root, 
                    text=texto, 
                    width=10, 
                    height=2, 
                    bg='green',  # Cambia el color de fondo a verde
                    command=self.mostrar_resultado
                )
            elif texto == 'C':
                boton = tk.Button(
                    self.root, 
                    text=texto, 
                    width=10, 
                    height=2, 
                    bg='red',  # Cambia el color de fondo a rojo
                    command=self.limpiar_pantalla
                )
            else:
                boton = tk.Button(
                    self.root, 
                    text=texto, 
                    width=10, 
                    height=2, 
                    command=lambda t=texto: self.agregar_a_pantalla(t)
                )

            boton.grid(row=fila, column=columna, padx=5, pady=5)

    def agregar_a_pantalla(self, valor):
        """Agrega el valor presionado a la pantalla y a la expresión."""
        self.calculadora.agregar(valor)
        self.pantalla.delete(0, tk.END)
        self.pantalla.insert(tk.END, self.calculadora.expresion)

    def limpiar_pantalla(self):
        """Limpia la pantalla y la expresión actual."""
        self.calculadora.limpiar()
        self.pantalla.delete(0, tk.END)

    def mostrar_resultado(self):
        """Muestra el resultado del cálculo en la pantalla."""
        resultado = self.calculadora.calcular()
        self.pantalla.delete(0, tk.END)
        self.pantalla.insert(tk.END, resultado)

# Programa principal
if __name__ == "__main__":
    root = tk.Tk()
    interfaz = InterfazCalculadora(root)
    root.mainloop()



```
