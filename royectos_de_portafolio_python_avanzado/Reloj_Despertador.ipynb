 RELOJ DESPERTADOR :
 
Este es un concepto interesante para un proyecto en Python. Las aplicaciones de reloj 
despertador son utilizadas por personas en todo el mundo. Para un desarrollador intermedio 
en Python, es un programa bastante sencillo en la Línea de Comando (CLI). Sin embargo, 
este proyecto no es tu típico reloj despertador. Puedes ingresar en realidad URLs de YouTube 

```python
import tkinter as tk
from tkinter import messagebox
import time
import webbrowser
from datetime import datetime

# Lista global para almacenar alarmas
alarmas = []

def agregar_alarma():
    """
    Función para agregar una alarma a la lista desde la GUI.
    """
    hora = entrada_hora.get()
    url = entrada_url.get()

    if not hora or not url:
        messagebox.showwarning("Advertencia", "Por favor, completa todos los campos.")
        return

    alarmas.append((hora, url))
    lista_alarmas.insert(tk.END, f"{hora} - {url}")
    messagebox.showinfo("Éxito", f"Alarma agregada para las {hora} con el enlace {url}")


def iniciar_reloj():
    """
    Función para iniciar el reloj despertador desde la GUI.
    """
    def verificar_alarmas():
        while alarmas:
            hora_actual = datetime.now().strftime("%H:%M")
            for hora, url in alarmas:
                if hora_actual == hora:
                    webbrowser.open(url)
                    alarmas.remove((hora, url))
                    lista_alarmas.delete(0, tk.END)
                    for h, u in alarmas:
                        lista_alarmas.insert(tk.END, f"{h} - {u}")
                    messagebox.showinfo("Alarma", f"¡Es la hora! Abriendo {url}")
            time.sleep(30)

    import threading
    hilo = threading.Thread(target=verificar_alarmas, daemon=True)
    hilo.start()
    messagebox.showinfo("Reloj activado", "El reloj despertador está activo.")


def eliminar_alarma():
    """
    Función para eliminar la alarma seleccionada en la lista.
    """
    seleccion = lista_alarmas.curselection()
    if seleccion:
        indice = seleccion[0]
        alarmas.pop(indice)
        lista_alarmas.delete(indice)
        messagebox.showinfo("Éxito", "Alarma eliminada correctamente.")
    else:
        messagebox.showwarning("Advertencia", "Por favor, selecciona una alarma para eliminar.")

# Crear ventana principal
root = tk.Tk()
root.title("Reloj Despertador con YouTube")
root.geometry("500x400")

# Widgets para agregar alarmas
tk.Label(root, text="Hora (HH:MM):", font=("Arial", 12)).pack(pady=5)
entrada_hora = tk.Entry(root, font=("Arial", 12))
entrada_hora.pack(pady=5)

tk.Label(root, text="URL de YouTube:", font=("Arial", 12)).pack(pady=5)
entrada_url = tk.Entry(root, font=("Arial", 12), width=40)
entrada_url.pack(pady=5)

boton_agregar = tk.Button(root, text="Agregar Alarma", font=("Arial", 12), command=agregar_alarma)
boton_agregar.pack(pady=10)

# Lista para mostrar alarmas configuradas
tk.Label(root, text="Alarmas Configuradas:", font=("Arial", 12)).pack(pady=5)
lista_alarmas = tk.Listbox(root, font=("Arial", 12), width=50, height=10)
lista_alarmas.pack(pady=5)

# Botones para iniciar y eliminar alarmas
boton_iniciar = tk.Button(root, text="Iniciar Reloj", font=("Arial", 12), command=iniciar_reloj)
boton_iniciar.pack(pady=10)

boton_eliminar = tk.Button(root, text="Eliminar Alarma", font=("Arial", 12), command=eliminar_alarma)
boton_eliminar.pack(pady=5)

# Ejecutar la aplicación
root.mainloop()
```
